```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from math import sqrt
```

MAI JOS AM INCERCAT SA VAD DACA EXISTA VREO LEGATURA INTRE NOTELE LUATE IN SESIUNILE 2-6 SI PUNCTAJUL LA TESTUL FINAL
X-VARIABILE INDEPENDENTE
Y-VARIABILE DEPENDETE

```python
predictori = ['Session 2','Session 3','Session 4','Session 5','Session 6']

X = notei_f[predictori]
y = notei_f['TOTAL\n(100 points)']

#IMPART DATELE IN SETURI DE ANTRENAMENT SI DE TESTARE. MARIMEA SETULUI DE TEST E DE 30%, IAR CEA
#A SETULUI DE ANTRENAMENT E DE 70%
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100)

print(X_train)
print (y_train)
```

ANTRENAREA ALGORITMULUI

```python
lm = LinearRegression()
model = lm.fit(X_train, y_train)

print('alfa=constanta = ', model.intercept_)
print('lista coef beta=', model.coef_)
```

ECUATIA CARE REZULTA ESTE:
f(x)=20.07+ session2*(4.75)+ 0.2*session3 + 3.06*session4+ (-0.9)*session5 +5.74*session6

FAC PREDICTII - FOLOSESC DATELE DE TEST SI VAD CAT DE PRECIS PREZICE ALGORITMUL VALORILE

```python
y_predDateTest = model.predict(X_test) 
tabel = pd.DataFrame({'Real Values':y_test, 'Predicted Values':y_predDateTest})
tabel
```

EVALUEZ MODELUL


AM CALCULAT R PATRAT(COEFICIENTUL DE DETERMINARE) CARE ARATA CAT DE BINE E ANTRENAT MODELUL -> varianta  DIN Y CARE ARATA CAT DE BINE POATE FI EXPLICAT X 

```python

r_patrat = r2_score(y_test, y_predDateTest)

print(r_patrat)
```

R^2 = 33% -> acuratetea modelului e scazuta. O ACURATETE BUNA AR FI DATA DE R>70%

```python

print("MAE: ",metrics.mean_absolute_error(y_test,y_predDateTest))
print("MSE: ", metrics.mean_squared_error(y_test,y_predDateTest))
print("RMSE: ", np.sqrt(metrics.mean_squared_error(y_test,y_predDateTest))) 
```

RMSE tb sa fie < 10 ->ALGORITMUL NU E ACURAT / SETUL DE DATE NU E BUN/E PREA MIC/ ATRIBUTELE FOLOSITE NU AU MARE CORELATIE INTRE ELE!!!!!!

X -> NOTELE DIN SESIUNILE 2-6, IAR Y -> REZULTATUL LA TESTUL FINAL: TRECUT/PICAT

```python
x=df[['Session 2','Session 3','Session 4','Session 5','Session 6']].values
y=df['Trecut/Picat'].values
```

```python
y
```

IMPORTURI

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

IMPARTIM SETUL DE DATE IN ANTRENAMENT SI TEST; 
RANDOM_STATE =10 -> TRAIN_TEST_SPLIT VA RETURNA ACELEASI REZULTATE PENTRU FIECARE EXECUTIE
1/3 -> DATE PENTRU TEST
2/3 -> DATE PENTRU ANTRENAMENT

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=1/3, shuffle=True, random_state=10)

#PRIMA DATA INCERCAM CU 5 VECINI
model = KNeighborsClassifier(n_neighbors=5)

#ANTRENAM MODELUL
model.fit(x_train, y_train)

```

DETERMINAM NUMARUL OPTIM DE VECINI

```python
k_interval = range(1,25)

scoruri = []

for k in k_interval:
    knn = KNeighborsClassifier(n_neighbors=k)  #instantiere
    knn.fit(x_train, y_train.ravel())               #antrenare
    y_prezis = knn.predict(x_test)
    scoruri.append(metrics.accuracy_score(y_test, y_prezis))
    
print(scoruri)
```

DIN GRAFICUL DE MAI JOS REIESE FAPTUL CA NUMARUL OPTIM DE VECINI AR FI UNDEVA INTRE 12 SI 15. AM ALES VALOAREA 12

```python
sns.lineplot(x=k_interval, y=scoruri)
```

```python
#antrenarea efectiva
knn = KNeighborsClassifier(n_neighbors=12)
knn.fit(x_train, y_train)

#FACEM PREDICTII PE SETUL DE TEST
y_pred= knn.predict(x_test)

print(y_train.reshape(1,-1))
print(y_pred)
#acestea sunt rezultatele prezicerii pe setul de test
```

MAI JOS EVALUAM ALGORITMUL

```python
print(metrics.accuracy_score(y_test, y_pred))
```

CELULA DE MAI JOS SEMINIFICA FAPTUL CA DIN 23 DE INTREGISTRARI DIN SETUL DE DATE, 15 AU FOST CLASIFICATE CORECT SI 8 INCORECT

```python

print(confusion_matrix(y_test, y_pred))
```

CELULA DE MAI JOS SEMNIFICA FAPTUL CA DIN CELE 23 DE INREGISTRARI DIN SETUL DE TEST, 64% AU FOST PREZISE CORECT IN CAZUL STUDENTILOR PICATI SI 67% IN CAZUL STUDENTILOR CARE AU LUAT EXAM FINAL

```python
print(classification_report(y_test, y_pred))
```

DACA AR FI SA INTRODUCEM NOI VALORI CARE SA FIE CLASIFICATE FACEM IN FELUL URMATOR 

```python
val_de_prezis=np.array([6,1,5,2,3]).reshape(1,-1)
y_pred=knn.predict(val_de_prezis)

#print(y_train)
print(y_pred)
```

# KMEANS

```python
import pandas as pd
note_finale=note_finale1
note_finale.head(10)
```

```python
X = note_finale.iloc[:, [1,16]].values
X
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1 , 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 10)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1, 11), wcss,marker='o')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 2, init = 'k-means++', random_state = 10)
y_kmeans = kmeans.fit_predict(X)
y_kmeans
```
